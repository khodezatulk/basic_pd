```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("/content/drive/MyDrive/Machine_learning/homeprices.csv")
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
rg =linear_model.LinearRegression()#create object
rg.fit(df[['area']],df.price)
```

```python
price=rg.predict(np.array(3300).reshape(-1, 1))
```
